```python
import pandas as pd

df = pd.read_csv("D:\FIFA2026-Player-Analytics\data\processed\cleaned_dataset.csv")

df.head()
```

```python
print(df.shape)

df.info()

df.describe()
```

```python
df["Position"].value_counts()

```

```python
import matplotlib.pyplot as plt

df["Position"].value_counts().plot(kind="bar")

plt.title("Players by Position")
plt.xlabel("Position")
plt.ylabel("Count")

plt.show()
```

```python
country_counts = (
    df["Country"]
    .value_counts()
    .head(20)
)

country_counts
```

```python
plt.figure(figsize=(12,5))

country_counts.plot(kind="bar")

plt.title("Top 20 Countries by Number of Players")
plt.xlabel("Country")
plt.ylabel("Players")

plt.show()
```

```python
import matplotlib.pyplot as plt

missing = df.isnull().sum()

missing = missing[missing > 0]

missing.sort_values().plot(kind="barh")

plt.title("Missing Values")

plt.show()
```

```python
goals = (
    df.groupby("Position")["Goals"]
      .sum()
      .sort_values(ascending=False)
)

goals
```

```python
plt.figure(figsize=(6,4))

goals.plot(kind="bar")

plt.title("Goals by Position")
plt.ylabel("Goals")

plt.show()
```

```python
top_scorers = (
    df.sort_values(
        "Goals",
        ascending=False
    )
    [["Player","Country","Goals"]]
    .head(15)
)

top_scorers
```

```python
plt.figure(figsize=(12,5))

plt.bar(
    top_scorers["Player"],
    top_scorers["Goals"]
)

plt.xticks(rotation=70)

plt.title("Top Goal Scorers")

plt.show()
```

```python
top_assists = (
    df.sort_values(
        "Assists",
        ascending=False
    )
    [["Player","Country","Assists"]]
    .head(15)
)

top_assists
```

```python
top_xg = (
    df.sort_values(
        "xG",
        ascending=False
    )
    [["Player","Country","xG"]]
    .head(15)
)

top_xg
```
