```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("D:\\FIFA2026-Player-Analytics\\data\\processed\\engineered_dataset.csv")

df.head()
```

```python
top_attack = (
    df.sort_values(
        "Attacking Impact Score",
        ascending=False
    )[
        [
            "Player",
            "Country",
            "Position",
            "Attacking Impact Score"
        ]
    ]
    .head(20)
)

top_attack
```

```python
plt.figure(figsize=(12,6))

plt.barh(
    top_attack["Player"],
    top_attack["Attacking Rating"]
)

plt.title("Top 20 Attacking Players")

plt.gca().invert_yaxis()

plt.show()
```
